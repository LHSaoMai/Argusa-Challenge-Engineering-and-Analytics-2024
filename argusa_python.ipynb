Compute the revenue

```python
# First, merge by product_id
merged = orders.merge(products, left_on='product_id', right_on='id', how='left')

# Then, filter for date range
orders_products = merged[
    (merged['order_date'] >= merged['start_date']) &
    (merged['order_date'] <= merged['end_date'])]
```

```python
orders_products.head()
```

```python
# 4. Count quantity ordered
order_summary = orders_products.groupby(['order_id','client_id','product_id','category_1', 'base_price','sale_price'], as_index=False).size()
order_summary.rename(columns={'size': 'quantity'}, inplace=True)
```

```python
# 4. Count quantity ordered
order_summary.head()
```

```python
returns.head()
```

```python
# count quantity returns
return_summary = returns.groupby(['order_id','product_id'], as_index=False).size()
return_summary.rename(columns={'size': 'quantity_returned'}, inplace=True)
```

```python
return_summary.head()
```

```python
final = order_summary.merge(return_summary, on=['order_id', 'product_id'], how='left')
```

```python
final['quantity_returned'] = final['quantity_returned'].fillna(0)
```

```python
final.head()
```

```python
final['total_quantity']=final['quantity']-final['quantity_returned']
```

```python
final['revenue'] = final['sale_price']*final['total_quantity']
```

Compute the margin

```python
final['margin']= (final['sale_price'] - final['base_price'])*final['quantity']
```

Compute sale

```python
final['sale'] = final['sale_price']*final['quantity']
```

```python
final.head()
```

Net profit

```python
final['net_profit'] = (final['sale_price']-final['base_price'])*(final['total_quantity'])
```

```python
final.head()
```

Net profit

```python
final['profit_ratio']=final['net_profit']/final['revenue']
```

```python
final.head()
```

Average basket

```python
average_basket = final['sale'].sum() / final['order_id'].nunique()
average_basket
```

Check point - global revenue

```python
final['revenue'].sum() # yes correct
```

```python
difference = (rating_mean - m).abs()

max_diff_product_id = difference.idxmax()
max_diff_value = difference.max()
```

```python
canton = final_client.groupby('address_canton')['net_profit'].sum()
canton.sort_values(ascending=False,inplace=True)

top_5_can=canton.head(5)
least_5 = canton.tail()
```

```python
canton.head()
```

```python
fig = plt.figure(figsize=(10,6))
axo= fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)


top_5_can.plot(kind='bar',ax=axo)

# Annotate each bar with the revenue value
for p in axo.patches:
    height = p.get_height()
    axo.annotate(f'{height:,.0f}',  # formats with commas and no decimals
                (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom', fontsize=10, fontweight='bold')

axo.set_title('Top 5 cantons by net profit')
axo.set_xlabel('Canton')



least_5.plot(kind='bar',ax=ax1)

for p in ax1.patches:
    height = p.get_height()
    ax1.annotate(f'{height:,.0f}',  # formats with commas and no decimals
                (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom', fontsize=10, fontweight='bold')


ax1.set_title('Least 5 cantons by revenue')
ax1.set_xlabel('Canton')

plt.tight_layout()

plt.savefig("question6.1.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
# fraction of net profit 
top_5_canton_names = top_5_can.index.tolist()
leat_5_canton_name = least_5.index.tolist()
top_5_canton_names
```

```python
final_client['canton_grouped'] = final_client['address_canton'].apply(
    lambda x: 'top group' if x in top_5_canton_names 
    else ('bottom group' if x in leat_5_canton_name else x)
)
```

```python
group_can=final_client.groupby('canton_grouped')['revenue'].sum()
group_can
```

```python
import matplotlib.cm as cm
colors = cm.get_cmap('tab20').colors

group_can.plot(kind='pie',
                figsize=(10, 6),
                autopct='%1.1f%%', # add in percentages
                startangle=90,     # start angle 90° (Africa)
                shadow=True,
                labels = None,# add shadow      
                pctdistance = 1.20,
                colors=colors)

plt.title('Contribution by canton',loc='center',y =1.12)
plt.axis('equal') # Sets the pie chart to look like a circle.
plt.legend(labels=group_can.index, loc='upper right', fontsize=7) 

plt.tight_layout()

plt.savefig("question6.2.png", dpi=300, bbox_inches='tight')

plt.show()
```
